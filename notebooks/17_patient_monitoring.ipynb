# Week 3 Day 3 — Patient Monitoring Dashboard

## ICU Early Warning Prediction System

The objective of this stage is to transform the dashboard into an interactive ICU monitoring tool.

The dashboard will allow clinicians to:

- monitor individual patients,
- review predicted sepsis risk,
- identify high-risk cases,
- prioritize critical patients,
- support clinical decision-making.

```python
import pandas as pd
from pathlib import Path

print("Libraries imported successfully.")
```

```python
PROJECT_ROOT = Path.cwd().parent

DASHBOARD_DATA_DIR = PROJECT_ROOT / "dashboard_data"

print("Dashboard data exists:", DASHBOARD_DATA_DIR.exists())
```

```python
patient_predictions = pd.read_csv(
    DASHBOARD_DATA_DIR / "dashboard_patient_predictions.csv"
)

top_10_patients = pd.read_csv(
    DASHBOARD_DATA_DIR / "dashboard_top_10_high_risk_patients.csv"
)

print("Patient predictions shape:", patient_predictions.shape)
print("Top 10 shape:", top_10_patients.shape)
```

```python
high_risk = patient_predictions[
    patient_predictions["Risk_Category"].isin(
        ["High Risk", "Critical Risk"]
    )
]

print("High/Critical patients:", len(high_risk))

high_risk.head()
```

## Clinical Interpretation

Displaying patient-level predictions allows clinicians to quickly identify individuals requiring closer monitoring.

Separating high-risk patients improves prioritization and may support earlier intervention.

The Top 10 highest-risk patients can assist ICU teams during clinical review and resource allocation.
