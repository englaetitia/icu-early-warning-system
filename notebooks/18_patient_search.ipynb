# Week 3 Day 4 — Patient Search and Filtering

## ICU Early Warning Prediction System

The objective of this stage is to improve dashboard usability by allowing clinicians to quickly locate specific patients and filter the monitoring view based on risk categories.

These functionalities support efficient clinical workflows by reducing the time required to identify patients requiring attention.

```python
import pandas as pd
from pathlib import Path

print("Libraries imported successfully.")
```

```python
PROJECT_ROOT = Path.cwd().parent

DASHBOARD_DATA_DIR = PROJECT_ROOT / "dashboard_data"

patient_predictions = pd.read_csv(
    DASHBOARD_DATA_DIR / "dashboard_patient_predictions.csv"
)

print("Patient predictions shape:", patient_predictions.shape)

patient_predictions.head()
```

```python
patient_predictions["Risk_Category"].value_counts()
```

```python
sample_patient = patient_predictions.iloc[0]["Patient_ID"]

searched_patient = patient_predictions[
    patient_predictions["Patient_ID"] == sample_patient
]

searched_patient
```

```python
critical_patients = patient_predictions[
    patient_predictions["Risk_Category"] == "Critical Risk"
]

print("Critical patients:", len(critical_patients))

critical_patients.head()
```

## Clinical Interpretation

Searching by Patient ID enables clinicians to rapidly retrieve individual patient information.

Filtering by risk category supports prioritization by focusing attention on high-risk populations.

These capabilities improve dashboard efficiency and simulate common functionalities found in hospital monitoring systems.
