# Week 3 Day 5 — Single Patient Dashboard

## ICU Early Warning Prediction System

The objective of this stage is to create a patient-centered dashboard view.

Clinicians will be able to select a patient and review their individual prediction results, risk profile, and clinical alert level.

This functionality simulates the patient overview pages commonly found in electronic health record systems and clinical decision support tools.

```python
import pandas as pd
from pathlib import Path

print("Libraries imported successfully.")
```

```python
PROJECT_ROOT = Path.cwd().parent

DASHBOARD_DATA_DIR = PROJECT_ROOT / "dashboard_data"

patient_predictions = pd.read_csv(
    DASHBOARD_DATA_DIR / "dashboard_patient_predictions.csv"
)

print("Shape:", patient_predictions.shape)

patient_predictions.head()
```

```python
selected_patient = patient_predictions.iloc[0]

selected_patient
```

```python
print("Patient ID:", selected_patient["Patient_ID"])
print("True Sepsis Label:", selected_patient["True_Sepsis_Label"])
print("Predicted Label:", selected_patient["Predicted_Sepsis_Label"])
print("Sepsis Probability:", selected_patient["Sepsis_Probability"])
print("Risk Score (%):", selected_patient["Risk_Score_Percent"])
print("Risk Category:", selected_patient["Risk_Category"])
print("Clinical Alert:", selected_patient["Clinical_Alert_Level"])
```

## Clinical Interpretation

Patient-level dashboards provide clinicians with concise summaries of model outputs.

Presenting individual predictions improves transparency and facilitates targeted clinical review.

These interfaces are common components of healthcare decision support systems.
