# Week 3 Day 2 — Dashboard Summary Metrics

## ICU Early Warning Prediction System

In Week 3 Day 1, the initial Streamlit dashboard was created.

Today, the objective is to connect the dashboard to real prediction outputs generated in Week 2.

The dashboard will display ICU-level summary metrics such as:

- total number of patients
- predicted sepsis cases
- high-risk and critical-risk patients
- average sepsis risk score
- risk category distribution

This transforms the dashboard from a static interface into a data-driven healthcare monitoring tool.

```python
import pandas as pd
from pathlib import Path

print("Libraries imported successfully.")
```

```python
PROJECT_ROOT = Path.cwd().parent

DASHBOARD_DATA_DIR = PROJECT_ROOT / "dashboard_data"

print("Dashboard data path:", DASHBOARD_DATA_DIR)
print("Dashboard data exists:", DASHBOARD_DATA_DIR.exists())
```

```python
summary_metrics = pd.read_csv(DASHBOARD_DATA_DIR / "dashboard_summary_metrics.csv")
risk_distribution = pd.read_csv(DASHBOARD_DATA_DIR / "dashboard_risk_distribution.csv")
patient_predictions = pd.read_csv(DASHBOARD_DATA_DIR / "dashboard_patient_predictions.csv")

print("Summary metrics shape:", summary_metrics.shape)
print("Risk distribution shape:", risk_distribution.shape)
print("Patient predictions shape:", patient_predictions.shape)
```

```python
summary_metrics
```

```python
def get_metric_value(metric_name):
    return summary_metrics.loc[
        summary_metrics["Metric"] == metric_name,
        "Value"
    ].values[0]


total_patients = get_metric_value("Total Patients")
true_sepsis_cases = get_metric_value("True Sepsis Cases")
predicted_sepsis_cases = get_metric_value("Predicted Sepsis Cases")
high_critical_patients = get_metric_value("High or Critical Risk Patients")
average_risk_score = get_metric_value("Average Risk Score (%)")

print("Total patients:", total_patients)
print("True sepsis cases:", true_sepsis_cases)
print("Predicted sepsis cases:", predicted_sepsis_cases)
print("High/Critical patients:", high_critical_patients)
print("Average risk score:", average_risk_score)
```

## Clinical Interpretation

The summary metrics provide a quick overview of ICU patient risk status.

These values are useful for a dashboard because clinicians need fast access to:

- number of monitored patients
- number of predicted septic patients
- number of high-priority cases
- average risk level across the ICU population

This gives the dashboard an ICU-level monitoring view.
